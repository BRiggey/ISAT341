```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
X = np.array([0, 1, 2])
Y = np.array([1, 1.5, 2])

# Model equation
def predict(w0, w1, x):
    return w0 + w1 * x

# Cases
cases = [(1, 0), (1, 1), (1, 2)]

# Plotting
plt.scatter(X, Y, color='red', label='Original data')
x_values = np.linspace(0, 2, 100)

for w0, w1 in cases:
    y_values = predict(w0, w1, x_values)
    label = f'Case: w0={w0}, w1={w1}'
    plt.plot(x_values, y_values, label=label)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression: Cases from Step a')
plt.legend()
plt.grid(True)
plt.show()
```
