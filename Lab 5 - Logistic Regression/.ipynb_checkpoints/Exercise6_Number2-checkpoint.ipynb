```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
z = np.linspace(-5, 5, 200)
s = sigmoid(z)
print(z.shape)
sigmoid(z)
plt.title("Sigmoid function")
plt.xlabel('Numbered value')
plt.ylabel('Mapped out Value between 0 and 1')

plt.plot(z,s)
```

```python
print('The probability of 0: {}'.format(sigmoid(0)))
```

```python
#3. 
thal = [3,6,3]
angina =[0,1,0]
slope = [2,3,1]
for i in range(3):
    z = (-1.482)*angina[i] - (0.670)*slope[i] - (0.604)*thal[i] + 4.756
    print('The Predicted Probability before rounding: {0:0.2f}'.format(sigmoid(z)))
    print('The Predicted Value for each row of values: {0:0.2f}'.format(z))
    print('The Class {0:0.2f}'.format(round(sigmoid(z))))
```

The middle value is the only value that is equal to class 0. The other two classes, the first and the third, are both predicted to be class 1.
